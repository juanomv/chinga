```python
import csv
import pandas as pd
# Lista para almacenar filas del archivo TSV
data = []

# Lee el archivo TSV
with open('ESTIMULOS POSITIVOS 22 Metrics.tsv', newline='') as tsvfile:
    reader = csv.reader(tsvfile, delimiter='\t')
    for row in reader:
        data.append(row)

df= pd.DataFrame(data)
```

```python
import pandas as pd

# Suponiendo que ya tienes tu DataFrame df
# Ordenar el DataFrame por el nombre del participante
df_ordenado = df.sort_values(by=[1])

# Imprimir el DataFrame ordenado
print(df_ordenado)
```

```python
# Creamos un diccionario para almacenar los DataFrames separados por participante
dataframes_por_participante = {}
df_participantes = df[df[1].str.contains("Participant", na=False)]

# Iteramos sobre cada grupo de participantes y almacenamos los DataFrames en el diccionario
for participante, grupo in df_participantes.groupby(1):
    dataframes_por_participante[participante] = grupo
```

```python
import os

# Iterar sobre los DataFrames separados por participante y escribir cada uno en un archivo Excel
for participante, df_participante in dataframes_por_participante.items():
    # Definir el nombre del directorio del participante
    directorio_participante = f"{participante}"
    # Crear el directorio si no existe
    if not os.path.exists(directorio_participante):
        os.makedirs(directorio_participante)
    # Definir el nombre del archivo Excel
    nombre_archivo = f"{directorio_participante}/{participante}.xlsx"
    
    # Escribir el DataFrame en el archivo Excel

    # Obtener la cabecera del DataFrame principal como una lista
    cabecera_lista = df.iloc[0].tolist()

    # Crear un DataFrame a partir de la cabecera como una fila
    cabecera_df = pd.DataFrame([cabecera_lista], columns=df.columns)

    # Concatenar la cabecera con el DataFrame del participante 1
    df_concatenado = pd.concat([cabecera_df, df_participante], ignore_index=True)

    df_concatenado.to_excel(nombre_archivo, index=False)

    print(f"Archivo Excel para {participante} creado correctamente en: {nombre_archivo}")
```

```python
import os

# Iterar sobre los DataFrames separados por participante y escribir cada uno en un archivo Excel
for participante, df_participante in dataframes_por_participante.items():
    # Buscar el segundo conjunto de números en el nombre del participante
    index = participante.find("Participante") + len("Participante")
    if index != -1:
        # Extraer el número del participante
        numero = participante[index:].strip()
        # Convertir el número a entero y sumarle 20
        try:
            numero_nuevo = int(numero) +20
            # Definir el nombre del directorio del participante
            directorio_participante = f"Participante {numero_nuevo}"
            # Crear el directorio si no existe
            if not os.path.exists(directorio_participante):
                os.makedirs(directorio_participante)
            # Definir el nombre del archivo Excel
            nombre_archivo = f"{directorio_participante}/participante {numero_nuevo}_posi.xlsx"
            # Escribir el DataFrame en el archivo Excel
            # Obtener la cabecera del DataFrame principal como una lista
            cabecera_lista = df.iloc[0].tolist()
            # Crear un DataFrame a partir de la cabecera como una fila
            cabecera_df = pd.DataFrame([cabecera_lista], columns=df.columns)
            # Concatenar la cabecera con el DataFrame del participante
            df_concatenado = pd.concat([cabecera_df, df_participante], ignore_index=True)
            df_concatenado.to_excel(nombre_archivo, index=False)
            print(f"Archivo Excel para {participante} creado correctamente en: {nombre_archivo}")
        except ValueError:
            print(f"Error: No se pudo convertir '{numero}' a un número entero.")
    else:
        print(f"Error: No se encontró un segundo conjunto de números en '{participante}'.")
```
